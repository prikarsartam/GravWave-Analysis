```python
def visualize(K,basis):
    num=0
    user=int(input("Enter the number of samples to visualize: "))
    print("\n")
    datastore = K[0:user] # which effectively means len(datastore)=user
    #define subplots
    
    if user%basis==0:
        fig, ax = plt.subplots(int(user/basis), basis, figsize=(12,10))
        fig.tight_layout()
        for item in datastore:
            
            data = datastore[num]                # define Data
            
            x = [ele.real for ele in data]  # extract real part
            
            y = [ele.imag for ele in data]  # extract imaginary part

            #create subplots
            ax[num%(int(user/basis)), num%basis].scatter(x, y, color='blue',alpha=1,s=1)
            ax[num%(int(user/basis)), num%basis].set_ylabel('Imaginary')
            ax[num%(int(user/basis)), num%basis].set_xlabel('Real')
            num=num+1

    else:
        print("Enter a number divisible by 3")
        

visualize(show(dataset['H1']['SFTs']),3)
```
